```python
# imports
import os
import pathlib
import shutil
import pandas as pd
import glob
```

```python
excluded = ["splitting_edited_test_edited.csv","splitting_edited_train_edited.csv","splitting_edited_val_edited.csv","mappings.csv","images_dice.csv"]
base_dir = "./data_snakemake"
for root,dirs,files in os.walk("data_snakemake/mappings"):
    # iterate through all the mappings files
    # 

    for file in files:
        if file not in excluded:
            # split using "." E.g mappings.csv
            file_stem = file.split(".")[0]
            mapping_inference_folder = os.path.join(root,file_stem)
            mapping_inference_folder_data = os.path.join(mapping_inference_folder,"data")
            mapping_inference_folder_display = os.path.join(mapping_inference_folder,"display")
            mapping_inference_folder_seg = os.path.join(mapping_inference_folder,"segmentations")
            mapping_inference_folder_weight = os.path.join(mapping_inference_folder,"weights")
            mapping_inference_file = os.path.join(root,file)
            # print(mapping_inference_folder)
            # create a folder that is the same name as the mappings file
            pathlib.Path(mapping_inference_folder).mkdir(parents=True,exist_ok=True)
            pathlib.Path(mapping_inference_folder_data).mkdir(parents=True,exist_ok=True)
            pathlib.Path(mapping_inference_folder_display).mkdir(parents=True,exist_ok=True)
            pathlib.Path(mapping_inference_folder_seg).mkdir(parents=True,exist_ok=True)
            pathlib.Path(mapping_inference_folder_weight).mkdir(parents=True,exist_ok=True)
            # read mappings file
            mappings = pd.read_csv(mapping_inference_file,index_col=False)
            # drop the Unnamed: 0 column
            mappings = mappings.drop(columns=mappings.columns[0])
            # input image
            sources = mappings["Sources"][:]
            # visualisation of the ground truth segmentation map
            masks = mappings["Masks"][:]
            # the ground truth segmentation map
            annotations = mappings["Annotations"][:]
            zipped = zip(sources,masks,annotations)
            for image,masks,label in zipped:
                # get input data name
                split_image = image.split("/")
                image_name = split_image[-2]+split_image[-1]
                # get display name
                display_name = image_name.replace(".tif",".png")
                # get label name
                split_label = label.split("/")
                label_name = split_label[-2]+split_label[-1]
                # get weight name
                weight_name = image_name.replace(".tif",".npy")
                # print(f"image:{image_name}")
                # print(f"label:{label_name}")
                # print(f"weight:{weight_name}")
            # move the ground truth,input image and weights of the files to infer from the main data_folder to the created_folder
                # result = glob.glob("./data_snakemake/*/*/1373012_1426_2121_12369_LNHP_0000.*")
                # print(result)
                image_source = glob.glob(f"{base_dir}/data/*/{image_name}")[0]
                # print(image_source)
                display_source = glob.glob(f"{base_dir}/display/*/{display_name}")[0]
                label_source = glob.glob(f"{base_dir}/segmentations/*/{label_name}")[0]
                weight_source = glob.glob(f"{base_dir}/weights/*/{weight_name}")[0]

                image_dest = f"{mapping_inference_folder_data}/{image_name}"
                display_dest = f"{mapping_inference_folder_display}/{display_name}"
                label_dest = f"{mapping_inference_folder_seg}/{label_name}"
                weight_dest = f"{mapping_inference_folder_weight}/{weight_name}"
                # print(image_dest)
                # print(label_dest)
                # print(weight_dest)
                # print("\n")
                print(image_source)
                print(label_source)
                print(weight_source)
                print("\n")
                # print(f"{base_dir}/data/*/{image_name}.*")
                shutil.copy(image_source,image_dest)
                shutil.copy(display_source,display_dest)
                shutil.copy(label_source,label_dest)
                shutil.copy(weight_source,weight_dest)

            # print(file)
        # else:
            # print(f"excluded {file}")
```
