### Script to convert to 0,1,2,3 array for mask -> caveat will be black 

```python
import torch
import torchvision.transforms as transforms
import os
import numpy as np

def create_0123(num, img_path, mask_path):
    if os.path.exists(img_path):
        pass
    else:
        os.mkdir(img_path)
    if os.path.exists(mask_path):
        pass
    else:
        os.mkdir(mask_path)
    for n in range(num):
        points = []
        for i in range(160):
            if i % 10 == 0:
                # lower_b = np.random.randint(-10, 310)
                x = i
                y = np.random.randint(150, 165)
                # upper_b = lower_b + 30
                points.append((x,y))
                
        Interpolated = []
        # linear interpolation 
        for i in range(len(points)-1):
        # for i in range(2):
            point1, point2 = points[i], points[i+1]
            count = 0
            for i in range(point1[0], point2[0]):
                step = (point2[1] - point1[1]) / ( point2[0] - point1[0] )
                y = int(point1[1] + (step*count))
                count += 1
                Interpolated.append((i,y))
        # generating the points vertically
        points = []
        for i in range(160):
            if i % 10 == 0:
                # lower_b = np.random.randint(-10, 310)
                x = i
                y = np.random.randint(100, 115)
                # upper_b = lower_b + 30
                points.append((x,y))
                
        Interpolated_1 = []
        # linear interpolation 
        for i in range(len(points)-1):
        # for i in range(2):
            point1, point2 = points[i], points[i+1]
            count = 0
            for i in range(point1[0], point2[0]):
                step = (point2[1] - point1[1]) / ( point2[0] - point1[0] )
                y = int(point1[1] + (step*count))
                count += 1
                Interpolated_1.append((i,y))

 
        mask = np.ones((200,300,3)).astype('uint8')
        mask[:,:,:] = 0
        mask[:,:150,:] = 1 # width
        mask[:100,:,:] = 2 # height
        mask[:100,:150,:] = 3

        # Horizontal wave 
        for i in range(0,150):
            mask[:Interpolated_1[i][1],i+1,:] = 3
        for i in range(150,300):
            mask[:Interpolated_1[i-150][1],i,:] = 2

        # Vertical wave
        for i in range(0,100):
            mask[i,100:Interpolated[i][1],:] = 3
        for i in range(100,200):
            mask[i,100:Interpolated[i-100][1],:] = 1
        for i in range(0,150):
            mask[:Interpolated_1[i][1],i:i+1,:] = 3
        
        from PIL import Image
        filename = '{1:1d}.jpg'.format(num,n)
        im = Image.fromarray(mask)
        im.save(os.path.join(mask_path, filename))

        # convert_pil = transforms.ToPILImage()
        # mask = convert_pil(mask)
        # filename = '{1:1d}.jpg'.format(num,n)
        # mask.save(os.path.join(mask_path,filename)
        # plt.savefig(os.path.join(mask_path, filename))

        image = np.ones((200,300,3)).astype('uint8')
        image[:,:,:] = [0,255,0]
        image[:,:150,:] = [0,0,255] # width
        image[:100,:,:] = [255,255,0] # height
        image[:100,:150,:] = [255,0,0]

        # Horizontal wave 
        for i in range(0,150):
            image[:Interpolated_1[i][1],i+1,:] = [255,0,0]
        for i in range(150,300):
            image[:Interpolated_1[i-150][1],i,:] = [255,255,0]

        # Vertical wave
        for i in range(0,100):
            image[i,100:Interpolated[i][1],:] = [255,0,0]
        for i in range(100,200):
            image[i,100:Interpolated[i-100][1],:] = [0,0,255]

        for i in range(0,150):
            image[:Interpolated_1[i][1],i:i+1,:] = [255,0,0]

        convert_pil = transforms.ToPILImage()
        image = convert_pil(image)
        current = num
        filename = '{1:1d}.jpg'.format(current,n)
        image.save(os.path.join(img_path,filename))
```

```python
create_0123(200, 'input/image', 'input/mask')
```
